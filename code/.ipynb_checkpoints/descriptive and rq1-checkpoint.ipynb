```python
import pandas as pd
import seaborn as sns
import string
import matplotlib.pyplot as plt

sns.set_context("talk")
sns.set_palette("colorblind") 
```

```python
# load data

all_annotations = pd.read_csv("../data/all_annotations.csv")
all_annotations = all_annotations.set_index('paper')
all_annotations.head()
```

### which contexts?

```python
contexts = all_annotations[['contexts:with_interactions', 'contexts:without_interactions',
       'contexts:content_analysis', 'contexts:writing',
       'contexts:training_data', 'contexts:advice', 'LLMs studied', 'link'
       ]]
```

```python
import numpy as np

# how many papers have multiple ones?
contexts_ = contexts[['contexts:with_interactions', 'contexts:without_interactions',
       'contexts:content_analysis', 'contexts:writing',
       'contexts:training_data', 'contexts:advice']]

# first combine social simulation into one
simulation = []
for _, row in contexts_.iterrows():
    if type(row['contexts:with_interactions']) != float or type(row['contexts:without_interactions']) != float:
        simulation.append(1)
    else:
        simulation.append(np.nan)
contexts_['contexts:simulation'] = simulation
contexts_ = contexts_[['contexts:simulation', 'contexts:content_analysis', 'contexts:writing',
       'contexts:training_data', 'contexts:advice']]

# check instances which have several contexts
contexts_["#_context"] = 5 - contexts_.isna().sum(axis = 1)
contexts_.groupby("#_context").size()
```

```python
contexts_[~contexts_["contexts:training_data"].isna()]
```

```python
df = pd.DataFrame(contexts_.iloc[2].isna()).reset_index()
```

```python
exact_contexts = []
for _, row in contexts_.iterrows():
    if row['#_context'] == 0:
        exact_contexts.append("generic")
    elif row['#_context'] == 1:
        df = pd.DataFrame(row.isna()).reset_index()
        context = df[~df[df.columns[1]]]['index'].values[0].split(':')[1]
        exact_contexts.append(context)
    else:
        exact_contexts.append("multi-context")
        
contexts_["exact_context"] = exact_contexts
```

### evaluation mechanism (open vs. closed)

```python
open_closed = measures[['eval_open', 'eval_closed', 'eval_other']]

open_closed_type = []
for _, row in open_closed.iterrows():
    if (row["eval_closed"]) == 1.0 and (row["eval_open"]) == 1.0:
        open_closed_type.append("both")
    elif (row["eval_closed"]) == 1.0 and type(row["eval_open"]) == float:
        open_closed_type.append("closed")
    elif (row['eval_open']) == 1.0 and type(row['eval_closed']) == float:
        open_closed_type.append("open")
    else:
        open_closed_type.append("other")

open_closed["Evaluation_Type"] = open_closed_type
open_closed.groupby('Evaluation_Type').size()
```

## which people?

```python
people = all_annotations[['target_pop', 'input', 'object', 'default', 'gender', 'race',
       'nationality', 'location', 'immigration', 'age', 'education',
       'political', 'disability', 'religion', 'class', 'sexuality', 'other',
       'Conclusion on Representativeness?']]
```

```python
demo_categories = ['gender', 'race', 'nationality',
                 'location', 'immigration', 'age', 'education',
                 'political', 'disability', 'religion', 'class',
                  'sexuality', 'other']
```

```python
input_output = people[["input", "object"]]

input_output_type = []
for _, row in input_output.iterrows():
    if type(row['input']) != float and type(row['object']) == float:
        input_output_type.append("input")
    elif type(row["object"]) != float and type(row["input"]) == float:
        input_output_type.append("object")
    elif type(row["object"]) != float and type(row["input"]) != float:
        input_output_type.append("both")
    else:
        input_output_type.append("none")

input_output["Representation_Type"] = input_output_type
```

## combine

```python
all_annotations = all_annotations.merge(open_closed[['Evaluation_Type']],
                                        on = 'paper').merge(input_output[['Representation_Type']],
                                                            on = 'paper')
all_annotations.head()
```

#### Have duplicate rows for papers with multiple contexts, instead of having a multi-context category

```python
exact_contexts = []
new_contexts_df = pd.DataFrame()
new_rows = []
for _, row in contexts_.iterrows():
    if row['#_context'] == 0:
        exact_contexts.append("generic")
        new_rows.append(row)
    elif row['#_context'] == 1:
        df = pd.DataFrame(row.isna()).reset_index()
        context = df[~df[df.columns[1]]]['index'].values[0].split(':')[1]
        exact_contexts.append(context)
        new_rows.append(row)
    else:
        df = pd.DataFrame(row.isna()).reset_index()
        multiple_contexts = df[~df[df.columns[1]]]['index'].values
        for i in range(row['#_context']):
            new_rows.append(row)
            exact_contexts.append(multiple_contexts[i].split(':')[1])
        
new_contexts_df = pd.DataFrame(new_rows)
new_contexts_df['context'] = exact_contexts
# new_contexts_df = new_contexts_df.reset_index()
new_contexts_df.index.name = 'paper'
new_contexts_df
```

```python
new_contexts_df.groupby('context').size()
```

```python
contexts_eval_people = new_contexts_df[['context']].merge(all_annotations, on = 'paper',
                                             how = 'outer')
contexts_eval_people.head()
```

```python
len(contexts_eval_people)
```

```python
contexts_eval_people.groupby('context').size().sort_values(ascending = False) / len(contexts_eval_people)
```

```python
# fig, axs = plt.subplots(2, 1)

contexts_eval_people = contexts_eval_people[contexts_eval_people['Representation_Type'] != 'none']
toplot = pd.DataFrame(contexts_eval_people.groupby(['context',
                                                    'Representation_Type']).size()).reset_index()
toplot.columns = ['context', 'Representation Type', 'Count']
ax1 = sns.catplot(
    toplot, kind="bar", sharex = True,
    x="Representation Type", y="Count", col="context",
    col_order = ['advice', 'simulation', 'generic', 'content_analysis',
                'writing'],
    height=3, aspect=1, palette=sns.color_palette("colorblind")
)
ax1.set_titles("{col_name}")

ax1.set_axis_labels(x_var="")
ax1.set_xticklabels(labels=["both", "impers.", "personal."],fontsize=11.5)
for ax in ax1.axes.flat:
    ax.set_title(ax.get_title().replace("_", " "), fontsize=15.5)
    ax.spines['bottom'].set_visible(False)
    ax.tick_params(axis='x', which='both', length=0)
    
ax1.figure.subplots_adjust(top=0.8);
# ax1.fig.suptitle('a) Context vs. Representation Type')
plt.tight_layout()
plt.savefig("../figures/context_representation.pdf", dpi = 200)
```

```python
toplot = pd.DataFrame(contexts_eval_people.groupby(['context',
                                                    'Evaluation_Type']).size()).reset_index()
toplot.columns = ['context', 'Eval Mechanism', 'Count']
ax2 = sns.catplot(
    toplot, kind="bar", sharex = True,
    x="Eval Mechanism", y="Count", col="context",
    col_order = ['advice', 'simulation', 'generic', 'content_analysis',
                'writing'],
    height=3, aspect=1, palette=sns.color_palette("colorblind")
)
ax2.set_titles("{col_name}")

ax2.set_axis_labels(x_var="")
ax2.set_xticklabels(fontsize=11.5)
for ax in ax2.axes.flat:
    ax.set_title(ax.get_title().replace("_", " "), fontsize=15)
    ax.spines['bottom'].set_visible(False)
    ax.tick_params(axis='x', which='both', length=0)

ax2.figure.subplots_adjust(top=0.8);
# ax2.fig.suptitle('b) Context vs. Evaluation Mechanism')
plt.tight_layout()
plt.savefig("../figures/context_evaluation.pdf", dpi = 200)
```

### RQ1: demographic heatmaps by context, eval mechanism, and representation type

```python
context_categories = new_contexts_df['context'].unique()
context_categories
```

```python
contexts_eval_people.columns
```

```python
# one for contexts
demo_contexts = {}
col_cat = {}
heatmap_context_df = pd.DataFrame()
for cat in context_categories:
    # print(cat)
    demo_contexts[cat] = contexts_eval_people[contexts_eval_people['context'] == cat]
    col_cat[cat] = pd.DataFrame(len(demo_contexts[cat]) - demo_contexts[cat][demo_categories].isna().sum().sort_values()).reset_index()
    col_cat[cat].columns = ['demographic', cat]
    if cat == 'simulation':
        heatmap_context_df = col_cat[cat]
    else:
        heatmap_context_df = heatmap_context_df.merge(col_cat[cat], on = 'demographic')
                
heatmap_context_df.set_index('demographic', inplace = True)
heatmap_context_df = heatmap_context_df.reindex(['gender', 'race', 'age', 'political', 'education', 'class', 'location',
       'religion', 'disability', 'nationality', 'immigration',
       'sexuality', 'other'])
heatmap_context_df = heatmap_context_df[['advice', 'simulation', 'content_analysis',
                'writing', 'generic', 'training_data']]
heatmap_context_df = heatmap_context_df / heatmap_context_df.sum()

```

```python
# one for input/output
demo_contexts = {}
col_cat = {}
heatmap_io_df = pd.DataFrame()
for cat in contexts_eval_people['Representation_Type'].unique():
    # print(cat)
    demo_contexts[cat] = contexts_eval_people[contexts_eval_people['Representation_Type'] == cat]
    col_cat[cat] = pd.DataFrame(len(demo_contexts[cat]) - demo_contexts[cat][demo_categories].isna().sum().sort_values()).reset_index()
    col_cat[cat].columns = ['demographic', cat]
    if cat == contexts_eval_people['Representation_Type'].unique()[0]:
        heatmap_io_df = col_cat[cat]
    else:
        heatmap_io_df = heatmap_io_df.merge(col_cat[cat], on = 'demographic')
        
heatmap_io_df.set_index('demographic', inplace = True)
heatmap_io_df = heatmap_io_df.reindex(['gender', 'race', 'age', 'political', 'education', 'class', 'location',
       'religion', 'disability', 'nationality', 'immigration',
       'sexuality', 'other'])

heatmap_io_df = heatmap_io_df / heatmap_io_df.sum()

```

```python
# one for eval mechanism
demo_contexts = {}
col_cat = {}
heatmap_eval_df = pd.DataFrame()
for cat in contexts_eval_people['Evaluation_Type'].unique():
    # print(cat)
    demo_contexts[cat] = contexts_eval_people[contexts_eval_people['Evaluation_Type'] == cat]
    col_cat[cat] = pd.DataFrame(len(demo_contexts[cat]) - demo_contexts[cat][demo_categories].isna().sum().sort_values()).reset_index()
    col_cat[cat].columns = ['demographic', cat]
    if cat == contexts_eval_people['Evaluation_Type'].unique()[0]:
        heatmap_eval_df = col_cat[cat]
    else:
        heatmap_eval_df = heatmap_eval_df.merge(col_cat[cat], on = 'demographic')

heatmap_eval_df.set_index('demographic', inplace = True)
heatmap_eval_df = heatmap_eval_df.reindex(['gender', 'race', 'age', 'political', 'education', 'class', 'location',
       'religion', 'disability', 'nationality', 'immigration',
       'sexuality', 'other'])

heatmap_eval_df = heatmap_eval_df / heatmap_eval_df.sum()

```

```python
heatmap_io_df.columns = ["both", "impers.", "personal."]
```

```python
f, ax = plt.subplots(1, 2, figsize=(16, 6.5), sharey = True)
sns.heatmap(heatmap_io_df, annot=True, fmt="0.2f", linewidths=.5,
            ax=ax[0], cmap="Blues")
sns.heatmap(heatmap_eval_df, annot=True, fmt="0.2f", linewidths=.5,
            ax=ax[1], cmap="Blues")
ax[0].set_xlabel('Persona Type')
ax[1].set_xlabel('Response Format')

plt.suptitle('Demographic Distribution across Persona Type and Evaluation Mechanism')
plt.tight_layout()
plt.savefig('../figures/demographic_descriptive_app.pdf', dpi = 200)

```
